# 2. Data Partitioning

### a) Describe training, validation and test dataset. 

Training, Validation and Test Dataset are dataset splits while training Machine Learning models.

### b) What is the purpose for each of these split?

(1) Training Dataset: The subset of data used to fit a model
(2) Validation Dataset: The subset of data used to to perform frequent evaluation over the model
(3) Test Dataset: The subset of data used to provide an unbiased evaluation of a final model.

### c) What is k-fold cross validation? 

K-Fold Cross Validation is a model validation technic, by performing multiple rounds of evaluation on a model. With the dataset reandomly splitted into groups, and use a single group as testing set, while the rest as training set. And repeat the evaluation by rotating each group's role.

### d) What is the advantage and disadvantage of k-fold CV compared to normal training/test/validation method?

(1) Advantages: More accurate estimation and used the data more efficiently
(2) Disadvantage: Slower and more complicated

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
df_target = df[['TargetB']]
df_features = df.drop(['TargetB'], axis=1)
```

```python
# convert categorial data

from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()
df_features['StatusCat96NK'] = label_encoder.fit_transform(df_features['StatusCat96NK'])
df_features['DemGender'] = label_encoder.fit_transform(df_features['DemGender'])
```

```python
# split the data into training & test sets
x_train, x_test, y_train, y_test = train_test_split(
    df_features.values, df_target.values, 
    stratify = df_target.values, random_state = 0
)
```

```python
# build the decision tree classifier and fit with training datasetabs
tree_classifier = DecisionTreeClassifier(random_state = 0)
tree_classifier.fit(x_train, y_train)
```

```python
# Check the accurary
print(
    'Training Set Accuracy: {:.4f}'. 
    format(tree_classifier.score(x_train, y_train))
)

print(
    'Test Set Accuracy: {:.4f}'.
    format(tree_classifier.score(x_test, y_test))
)
```

- 100% accuracy for Training Set
- 51.69% accuracy for Test set
- Yes, over fitted
