```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
df = pd.read_csv('lab2.csv')
```

```python
df.head()
```

```python
df['SERVICE'] = df['SERVICE'].apply(lambda x: [x])
groupped_df = df.groupby(['ACCOUNT'])['SERVICE'].apply(sum).reset_index()
```

```python
groupped_df
```

```python
association_rules = apriori(groupped_df['SERVICE'], min_support=0.05)  
association_results = list(association_rules)
```

```python
association_results[0]
```

```python
rules = list()
support =list()
lift = list()
confidence=list()

for item in association_results:
    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    
    if len(items) == 1:
        rules.append("{} -> " + items[0])
    else:
        rules.append(items[0] + " -> " + items[1])
    support.append(str(item[1]))
    lift.append(str(item[2][0][2]))
    confidence.append(str(item[2][0][3]))
    
results = pd.DataFrame(
    list(zip(rules, support, lift, confidence)), 
    columns =['rules', 'support', 'lift', 'confidence']
) 
```

```python
results.sort_values(by='lift', ascending = False)
```
