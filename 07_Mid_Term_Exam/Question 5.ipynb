```python
import numpy as np 
import pandas as pd
from sklearn import preprocessing
```

```python
# load data from csv

data_directory = 'data/'

q4_transaction_csv = data_directory + 'q4_transactions.csv'
transaction_csv = data_directory + 'q5_transactions.csv'

transactions = pd.read_csv(q4_transaction_csv, na_values=['None'], index_col=0)
```

```python
transactions
```

```python
## Normalization

def normalize(input_column):
    # Create x, where x the 'scores' column's values as floats
    x = input_column.values.astype(float)

    # Create a minimum and maximum processor object
    min_max_scaler = preprocessing.MinMaxScaler()

    # Create an object to transform the data to fit minmax processor
    x_scaled = min_max_scaler.fit_transform(x)

    # Run the normalizer on the dataframe
    df_normalized = pd.DataFrame(x_scaled)
    
    return(df_normalized)


transactions[['area_sqft']] = normalize(transactions[['area_sqft']])
transactions[['bedrooms']] = normalize(transactions[['bedrooms']])
transactions[['median_psf']] = normalize(transactions[['median_psf']])
transactions[['median_price']] = normalize(transactions[['median_price']])
transactions[['nearby_poi_education']] = normalize(transactions[['nearby_poi_education']])
transactions[['nearby_poi_transportation']] = normalize(transactions[['nearby_poi_transportation']])
transactions[['nearby_poi_healthcare']] = normalize(transactions[['nearby_poi_healthcare']])

transactions
```

```python
transactions.to_csv(transaction_csv)
```
